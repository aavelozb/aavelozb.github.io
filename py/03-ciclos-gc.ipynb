### Iterar n veces (Contando de 0 a n-1)


````python
n = int( input('Cantidad de veces: ') )
i = 0
while i < n:
    print(i)
    i += 1
````

### Contando de 1 a n

````python
n = int(input('Cantidad de veces: '))
i = 1
while i <= n:
    print(i)
    i += 1
````

### Iterar hasta que se ingrese cierto valor

````python
n = int(input('Ingrese número (0 para terminar): '))
while n != 0:
    n = int(input('Ingrese número (0 para terminar): '))
````

### Iterar hasta que se cumpla una condicion particular

````python
suma = 0
while suma < 100:
    n = int(input('Ingrese número: '))
    suma += n
````

### Uso de banderas para terminar un ciclo

````python
n = int(input("Ingrese número a evaluar: "))
primo = True
divisor = 2

while primo and divisor <= n-1:
    if n%divisor == 0:
        primo = False
    divisor += 1

if primo:
    print (n, ' es primo')
else:
    print (n, ' es compuesto')
````

### Patrón de conteo

````python
n= int(input('Cantidad de números: '))
contador = 0
i = 0
while i < n:
    if i%2 == 0:
        contador += 1
    i += 1
print (contador, 'pares')
````

### Patrón para encontrar el mayor

````python
n= int(input('Cantidad de personas: '))
mayor = 0 # Alternativamente mayor float('-Inf')
i = 1
while i <= n:
    edad = int(input('Edad: '))
    if edad > mayor:
        mayor = edad
    i += 1
print('La mayor edad es:', mayor)
````

````python
n = int(input('Cantidad de personas: '))
mayor = int(input('Edad: ')) # Leer el primero antes del ciclo
i = 1
while i < n:
    edad = int(input('Edad: '))
    if edad > mayor:
        mayor = edad
    i += 1
print ('La mayor edad es:', mayor)
````

### Patrón para encontrar el menor

````python
n = int(input('Cantidad de personas: '))
menor = 1000 #Alternativamente mayor float('Inf')
i = 1
while i <= n:
    edad = int(input('Edad: '))
    if edad < menor:
        menor = edad
    i += 1
print('La menor edad es:', menor)
````

### Patrón de acumulación (suma)

````python
n = int(input('n: '))
suma = 0
i = 1
while i <= n:
    suma += i
    i += 1
print('La suma es:', suma)
````

### Patrón de acumulación (multiplicación)

````python
n = int(input('n: '))
factorial = 1
i = 1
while i <= n:
    factorial *=i
    i +=1
print('El factorial de', n,'es:', factorial)
````

### Patrón para calcular promedio (combinación de los patrones de conteo y suma)

````python
num = int(input('Ingrese número (O para terminar): '))
suma = num
n = 1
while num != 0:
    num = int(input('Ingrese número (0 para terminar): '))
    if num != 0:
        suma += num
        n += 1
promedio = suma / n
print(promedio)
````

### Ciclo anidado

````python
a = int(input("Ingrese limite inferior del rango: "))
b = int(input("Ingrese limite superior del rango: "))
candidato = a
while candidato<=b:
    # Evaluar si candidato es primo
    primo = True
    divisor = 2
    while primo and divisor <= candidato//2:
        if candidato%divisor == 0:
            primo = False
        divisor += 1
    if primo:
        print(candidato)
    candidato += 1
````
